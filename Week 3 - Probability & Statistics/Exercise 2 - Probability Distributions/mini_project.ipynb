## Mini Project 1 — Visualisasi Berbagai Distribusi

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, poisson, uniform

# gaussian distribution
x = np.linspace(-4, 4, 100)
plt.plot(x, norm.pdf(x, loc=0, scale=1), label='Normal Distribution')

#  binomial distribution
n, p = 10, 0.5
x_bin = np.arange(0, n+1)
plt.bar(x_bin, binom.pmf(x_bin, n, p), alpha=0.5, label='Binomial Distribution')

# poisson distribution
lambda_number = 3
x_pois = np.arange(0, 10)
plt.bar(x_pois, poisson.pmf(x_pois, lambda_number), alpha=0.5, label='Poisson Distribution')

# uniform distribution
x_uni = np.linspace(0, 1, 100)
plt.plot(x_uni, uniform.pdf(x_uni, loc=0, scale=1), label="Uniform (0–1)", color='red')

plt.title("Perbandingan Berbagai Distribusi Probabilitas")
plt.legend()
plt.xlabel("Nilai (x)")
plt.ylabel("Probabilitas")
plt.show()
```

## Mini Project 2 — Analisis Skewness & Kurtosis (Dataset Nyata)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

# pilih salah satu fitur
feature = df['sepal_length']

# hitung skewness dan kurtosis
skewness = skew(feature)
outliers = kurtosis(feature)

print(f"Skewness: {skewness}")
print(f"Kurtosis (Outliers): {outliers}")

sns.histplot(feature, kde=True)
plt.title("Distribusi Sepal Length (Iris Dataset)")
plt.show()
```
