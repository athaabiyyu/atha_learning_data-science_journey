## Mini Project 1: Confidence Interval dari Data Kecil

```python
import numpy as np
from scipy.stats import t

data = np.array([12, 14, 15, 16, 17, 18, 19])

# hitung mean dan standar deviasi sampel
mean = np.mean(data)
std_dev = np.std(data, ddof=1)
n = len(data)

# hitung t-score untuk 95% confidence interval
t_score = t.ppf(0.975, df = n-1)

# hitung margin erorr
margin_error = t_score * std_dev / np.sqrt(n)

# hitung CI
CI = (mean - margin_error, mean + margin_error)

print("Mean:", round(mean, 2))
print("95% Confidence Interval:", CI)
```

## Mini Project 2: Confidence Interval dari Data Besar

```python
import numpy as np
from scipy.stats import norm

sample = np.random.normal(loc=50, scale=10, size=100)

# hitung mean, std, dan ukuran sampel
mean = np.mean(sample)
std = np.std(sample, ddof=1)
n = len(sample)

# hitung z-score untuk 95% confidence interval
z_score = norm.ppf(0.975)

# hitung margin erorr
margin_error = z_score * std / np.sqrt(n)

# hitung CI
CI = (mean - margin_error, mean + margin_error)

print("Mean:", round(mean, 2))
print("95% Confidence Interval:", CI)
```

## Mini Project 3: Confidence Interval dari Dataset Nyata

```python
import numpy as np
import pandas as pd
from scipy.stats import norm

# Load data
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

# ambil 30 sample acak dari kolom 'sepal_length'
sample_sepal_length = df['sepal_length'].sample(n=30, random_state=42)

# hitung mean, std, dan ukuran sampel
mean = np.mean(sample_sepal_length)
std = np.std(sample_sepal_length, ddof=1)
n = len(sample_sepal_length)

# hitung z-score untuk 95% confidence interval
z_score = norm.ppf(0.975)

# hitung margin erorr
margin_error = z_score * std / np.sqrt(n)

# hitung CI
CI = (mean - margin_error, mean + margin_error)

print("Mean:", round(mean, 2))
print("95% Confidence Interval:", CI)
```
