##### Exercise 1 : Perform a Hyphotesis Test

```python
import numpy as np
from scipy.stats import ttest_1samp

# sample data 
data = [12, 14, 15, 14, 13, 16, 14, 15, 13, 14]

# null hypothesis mean
mean_h0 = 14

# perform t-test
t_stat, p_value = ttest_1samp(data, mean_h0)
print(f"T-statistic: {t_stat}, P-value: {p_value}")

if p_value < 0.05:
       print("Tolak H0, ada bukti yang cukup untuk menyatakan bahwa rata-rata populasi bukan 14")
else : 
       print("Gagal tolak H0, tidak ada bukti yang cukup untuk menyatakan bahwa rata-rata populasi bukan 14")  
```

##### Exercise 2 : Two Samples T-test

```python
from scipy.stats import ttest_ind

# sample data
data1 = [12, 14, 15, 14, 13, 16, 14, 15, 13, 14]
data2 = [22, 24, 19, 23, 25, 21, 20, 22, 24, 23]

# perform two-sample t-test
t_stat, p_value = ttest_ind(data1, data2)        
print(f"T-statistic: {t_stat}, P-value: {p_value}")

if p_value < 0.05:
       print("Tolak H0, ada bukti yang cukup untuk menyatakan bahwa rata-rata populasi kedua grup berbeda")
else : 
       print("Gagal tolak H0, tidak ada bukti yang cukup untuk menyatakan bahwa rata-rata populasi kedua grup berbeda")       
       
```

##### Additional Practice : Perform a z-test for large sample sizes

```python
import numpy as np
from scipy.stats import norm

# sample data
np.random.seed(42)
data = np.random.normal(loc=52, scale=10, size=1000)

# h0
mean_h0 = 50
std_h0 = 10
n = len(data)

# calculate z-score
mean_sample = np.mean(data)
z_score = (mean_sample - mean_h0) / (std_h0 / np.sqrt(n))

# hitung p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Z-score: {z_score}, P-value: {p_value}")
if p_value < 0.05:
       print("Tolak H0, ada bukti yang cukup untuk menyatakan bahwa rata-rata populasi bukan 50")
       
else : 
       print("Gagal tolak H0, tidak ada bukti yang cukup untuk menyatakan bahwa rata-rata populasi bukan 50")
```

##### Additional Practice : User the iris dataset to test if the mean sepal length differs between two spices

```python
import pandas as pd
from scipy.stats import ttest_ind

# load iris dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

# ambil data sepal_length untuk dua species
setosa = df[df['species'] == 'setosa']['sepal_length']
versicolor = df[df['species'] == 'versicolor']['sepal_length']

# uji t dua sampel independen
t_stat, p_value = ttest_ind(setosa, versicolor, equal_var=False)  # equal_var=False jika variansi berbeda

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# keputusan
if p_value < 0.05:
    print("Tolak H0 → Rata-rata sepal length berbeda")
else:
    print("Gagal tolak H0 → Rata-rata sepal length tidak berbeda")

```
