### Mini Project 1 : Simulasi Dadu

```python
import numpy as np

# lempar dadu sebanyak 1000 kali
data = np.random.randint(1, 7, size = 1000)

# hitung peluang genap
p_event = np.sum(data % 2 == 0) / len(data)
print("Peluang muncul angka genap:", p_event)

# hitung peluang muncul angka lebih dari 4
p_greater4 = np.sum(data>4) / len(data)
print("Peluang muncul angka lebih dari 4:", p_greater4)
```

## Mini Project 2 : Visualisasi PMF & PDF

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# visualisasi PMF - Dadu
data_pmf = np.array([1, 2, 3, 4, 5, 6])
pmf = np.ones(6)/6

plt.bar(data_pmf, pmf, color='blue', alpha=0.7)
plt.title('Probability Mass Function (PMF) of a Fair Die')
plt.xlabel('Hasil Dadu')
plt.ylabel('Probability')
plt.show()

# visualisasi PDF - Distribusi Kontinu 0 sampai 1
data_pdf = np.linspace(0, 1, 100)
pdf = uniform.pdf(data_pdf, loc=0, scale=1)

plt.plot(data_pdf, pdf, color='red')
plt.title("PDF - Uniform Distribution (0â€“1)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
