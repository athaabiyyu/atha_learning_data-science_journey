## Mini Project 1 — Pearson & Spearman Correlation

```python
# Mini Project 1: Pearson & Spearman Correlation
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, spearmanr
import seaborn as sns

# --- Example 1: Simple synthetic data ---
x = np.array([1,2,3,4,5])
y = np.array([2,4,5,8,10])

r_pearson, p_pearson = pearsonr(x, y)
r_spearman, p_spearman = spearmanr(x, y)

print("SYNTHETIC DATA")
print(f"Pearson r = {r_pearson:.4f}, p = {p_pearson:.4f}")
print(f"Spearman rho = {r_spearman:.4f}, p = {p_spearman:.4f}\n")

# Interpretation helper
def interpret(r):
    mag = abs(r)
    if mag >= 0.8: return "very strong"
    if mag >= 0.6: return "strong"
    if mag >= 0.4: return "moderate"
    if mag >= 0.2: return "weak"
    return "very weak"

print("Interpretation (synthetic):")
print(f"Pearson: {interpret(r_pearson)} {'positive' if r_pearson>0 else 'negative' if r_pearson<0 else 'no'} correlation")
print(f"Spearman: {interpret(r_spearman)} {'positive' if r_spearman>0 else 'negative' if r_spearman<0 else 'no'} correlation\n")

# --- Example 2: Iris dataset (numeric features only) ---
iris = sns.load_dataset('iris')
iris_numeric = iris.drop(columns=['species'])
print("IRIS DATASET: pairwise Pearson correlation (head):")
print(iris_numeric.corr().round(3))
```

## Mini Project 2 - Linear Regression + Visualisasi (step-by-step)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# generate data
np.random.seed(42)
X = np.random.rand(100, 1) * 10                # fitur X
true_slope = 3.0
true_intercept = 5.0
noise = np.random.randn(100, 1) * 2.0     # noise
y = true_intercept + true_slope * X + noise

# fit model
model = LinearRegression()
model.fit(X,y)
slope = model.coef_[0,0]
intercept = model.intercept_[0]
r2 = model.score(X, y)

print("Linear Regression Results:")
print(f"Slope (b1): {slope:.4f}")
print(f"Intercept (b0): {intercept:.4f}")
print(f"R-squared: {r2:.4f}")

# --- Visualize ---
plt.figure(figsize=(8,5))
plt.scatter(X, y, alpha=0.6, label='Data')
x_line = np.linspace(X.min(), X.max(), 100).reshape(-1,1)
y_line = model.predict(x_line)
plt.plot(x_line, y_line, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

## Mini Project 3 — Heatmap Correlation (Iris dataset)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Mini Project 3: Heatmap Correlation (Iris)
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

iris = sns.load_dataset('iris')
df = iris.drop(columns=['species'])  # numeric only
corr = df.corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Feature Correlation (Iris)')
plt.show()

print("\nCorrelation matrix:\n", corr.round(3))
```

## Mini Project 4 — Pearson & Spearman Correlation Analysis

```python
import numpy as np
from scipy.stats import pearsonr, spearmanr

# generate data
np.random.seed(42)

X = np.arange(1, 21)
Y = X * 2 + np.random.randint(-5, 5, size=20)

# hitung pearson dan spearman
r_pearson, p_pearson = pearsonr(X, Y)
r_spearman, p_spearman = spearmanr(X, Y)

# tampilkan hasil dan interpretasi
print("Pearson correlation:", round(r_pearson, 3), "| p-value =", round(p_pearson, 4))
print("Spearman correlation:", round(r_spearman, 3), "| p-value =", round(p_spearman, 4))
```

## Mini Project 5 - Simple Linear Regression + Visualisasi

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# buat data linear + noise
np.random.seed(42)
X = np.arange(1, 21).reshape(-1,1)
Y = 3 * X + np.random.randint(-10, 10, size=(20,1))

# latih model
model = LinearRegression()
model.fit(X,Y)

# prediksi dan ambil paramter model
Y_pred = model.predict(X)

slope = model.coef_[0][0]
intercept = model.intercept_[0]
r2 = model.score(X, Y)

print("Slope :", slope)
print("Intercept : ", intercept)
print("R2 : ", r2)

# visualisasi
plt.scatter(X, Y, label="Data")
plt.plot(X, Y_pred, label="Regression Line")
plt.title("Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
