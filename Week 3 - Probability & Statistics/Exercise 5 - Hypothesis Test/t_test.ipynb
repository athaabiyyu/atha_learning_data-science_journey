```python
import numpy as np
from scipy.stats import ttest_1samp, ttest_ind, ttest_rel

# data sample
data1 = np.array([1, 2, 3, 4, 5, 6])
data2 = np.array([7, 8, 9, 10, 11, 12])

# ---------------------------
# 1. ONE-SAMPLE T TEST
# ---------------------------
# mean = 3
hipothesis0 = 3

# hitung t-statistic dan p-value
t_stat1, p_value1 = ttest_1samp(data1, hipothesis0)

# menampilkan hasil
print("One-Sample T Test:")
print(f"T-statistic: {t_stat1}, P-value: {p_value1}")

alpha = 0.05
if p_value1 < alpha:
       print("Tolak hipotesis nol: rata-rata data1 berbeda secara signifikan dari 3")
else   :
       print("Gagal tolak hipotesis nol: rata-rata data1 tidak berbeda secara signifikan dari 3")
# ---------------------------
print()

# ---------------------------
# 2. TWO-SAMPLE T TEST
# ---------------------------
# hitung t-statistic dan p-value
t_stat2, p_value2 = ttest_ind(data1, data2)

# menampilkan hasil
print("Two-Sample T Test:")
print(f"T-statistic: {t_stat2}, P-value: {p_value2}")

alpha = 0.05
if p_value2 < alpha:
       print("Tolak hipotesis nol: rata-rata data1 dan data2 berbeda secara signifikan")
else   :
       print("Gagal tolak hipotesis nol: rata-rata data1 dan data2 tidak berbeda secara signifikan")
# ---------------------------
print()

# ---------------------------
# 3. PAIRED T TEST
# ---------------------------
pre_test  = np.array([12, 14, 15, 16, 17])
post_test = np.array([13, 14, 16, 17, 18])

t_stat_3, p_val_3 = ttest_rel(pre_test, post_test)

print("Paired T Test:")
print("T-Statistic:", t_stat_3)
print("P-Value:", p_val_3)

# tampilkan hasil uji hipotesis
alpha = 0.05
if p_val_3 < alpha:
       print("Tolak hipotesis nol: ada perbedaan signifikan antara pre-test dan post-test")
else:
       print("Gagal tolak hipotesis nol: tidak ada perbedaan signifikan antara pre-test dan post-test")
```
