```python
from scipy.stats import chi2_contingency

# contingency table (2x3)
data = [
    [50, 30, 20],
    [30, 40, 30]
]

chi2, p, dof, expected = chi2_contingency(data)

print("=== CHI SQUARE TEST ===")
print("Chi-Square Statistic:", chi2)
print("P-Value:", p)
print("Expected Frequencies:\n", expected)

alpha = 0.05
if p < alpha:
       print("Tolak hipotesis nol: ada asosiasi signifikan antara dua variabel kategori")
       
else:
       print("Gagal tolak hipotesis nol: tidak ada asosiasi signifikan antara dua variabel kategori")
```
