## Mini Project 1: Menguji apakah rata - rata berbeda dari angka tertentu

```python
import numpy as np
from scipy.stats import ttest_1samp

data = np.array([12, 14, 15, 16, 17, 18, 19])

# H0 : mean = 15
population_mean = 15.8

# hitung t-statistic dan p-value
t_stat, p_value = ttest_1samp(data, population_mean)

print("T-statistic:", t_stat)
print("P-value:", p_value)
print("Mean dari sampel:", np.mean(data))

# ambil keputusan
alpha = 0.05

if p_value < alpha:
       print("Tolak H0: Rata-rata berbeda dari", population_mean)
else:
       print("Gagal tolak H0: Rata-rata tidak berbeda dari", population_mean)
```

## Mini Project 2 : Two Sample - T-Test

```python
import numpy as np
from scipy.stats import ttest_ind

# data sampel
group1 = np.array([12, 14, 15, 16, 17, 18, 19])
group2 = np.array([11, 13, 14, 15, 16, 17, 18])

# H0 : mean_group1 = mean_group2
# hitung t-statistic dan p-value
t_stat, p_value = ttest_ind(group1, group2)
print("T-statistic:", t_stat)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
       print("Tolak H0: Rata-rata kedua grup berbeda")
else:
       print("Gagal tolak H0: Rata-rata kedua grup tidak berbeda")
```

## Mini Project: Apakah Minuman Energi Meningkatkan Fokus?

```python
import numpy as np
from scipy.stats import ttest_rel

pre = [60, 62, 58, 55, 64, 59, 61, 57, 63, 60]
post = [65, 66, 62, 60, 70, 63, 64, 59, 67, 65]

# H0 : mean_pre = mean_post
# hitung t-statistic dan p-value
t_stat, p_value = ttest_rel(pre, post)

# karena kita ingin menguji apakah ada peningkatan, kita gunakan one-tailed test
p_value_one_tailed = p_value / 2

print("T-statistic:", t_stat)
print("One-tailed P-value:", p_value_one_tailed)

alpha = 0.05
if p_value_one_tailed < alpha and t_stat > 0:
       print("Tolak H0: Ada peningkatan skor setelah pelatihan")
else :
       print("Gagal tolak H0: Belum ada bukti signifikan bahwa minuman energi meningkatkan fokus.")
```
