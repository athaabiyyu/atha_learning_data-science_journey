```python
import numpy as np
from scipy import stats

# Data nilai dua kelas
kelas_A = [78, 82, 88, 90, 75, 80]
kelas_B = [70, 72, 78, 74, 69, 75]

# hitung mean, variance dan standard deviation
mean_A = np.mean(kelas_A)
var_A = np.var(kelas_A)
std_A = np.std(kelas_A)

print(f"Kelas A \nMean: {mean_A}, \nVariance: {var_A}, \nStandard Deviation: {std_A}")

# lakukan t test
t_stat, p_value = stats.ttest_ind(kelas_A, kelas_B)
print(f"T-statistic: {t_stat}, P-value: {p_value}")

# interpretasi hasil
alpha = 0.05
if p_value < alpha:
    print("Tolak H0: Ada perbedaan signifikan antara kedua kelas.")
else:
    print("Gagal tolak H0: Tidak ada perbedaan signifikan antara kedua kelas.")
```
