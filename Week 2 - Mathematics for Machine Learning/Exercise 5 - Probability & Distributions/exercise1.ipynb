```python
# contoh PDF
import matplotlib.pyplot as plt
from scipy.stats import norm

# distribusi tinggi badan (mean = 170 cm, std = 10 cm)
x = np.linspace(140, 200, 1000)
pdf = norm.pdf(x, loc=170, scale=10)

plt.plot(x, pdf)
plt.title("PDF Tinggi Badan (Distribusi Normal)")
plt.xlabel("Tinggi Badan (cm)")
plt.ylabel("Probabilitas")
plt.show()
```
