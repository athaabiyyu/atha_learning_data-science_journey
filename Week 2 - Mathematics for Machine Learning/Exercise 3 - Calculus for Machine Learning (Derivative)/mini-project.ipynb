```python
# derivative dasar

import sympy as sp

x = sp.symbols('x')
f = x**2 + 3*x + 5
derivative = sp.diff(f,x)

print("Fungsi: ", f)
print("Turunan: ", derivative)
```

```python
# gradient fungsi 2 variabel

x, y = sp.symbols('x y')
f = x**2 + y**2 - 4*x*y

gradient_x = sp.diff(f, x)
gradient_y = sp.diff(f, y)

print("Fungsi: ", f)
print("Gradient terhadap x: ", gradient_x)
print("Gradient terhadap y: ", gradient_y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi gradient descent
def gradient_descent(x, y, theta, alpha, iterations):
    m = len(y)
    history = []  # untuk menyimpan theta setiap iterasi
    for _ in range(iterations):
        predictions = x.dot(theta)
        error = predictions - y
        gradients = (1/m) * x.T.dot(error)
        theta = theta - alpha * gradients
        history.append(theta.copy())
    return theta, np.array(history)

# Data sederhana
X = np.array([1, 2, 3])
y = np.array([2, 2.5, 3.5])

# Tambahkan kolom 1 untuk bias term
x = np.c_[np.ones(X.shape[0]), X]

# Nilai awal parameter
theta = np.array([0.1, 0.1])

# Jalankan gradient descent
theta_opt, history = gradient_descent(x, y, theta, alpha=0.1, iterations=50)

# Plot hasil
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='red', label='Data Asli')

# Plot beberapa garis selama proses training
for i in range(0, len(history), 5):  # ambil setiap 5 iterasi biar gak terlalu padat
    theta_i = history[i]
    plt.plot(X, x.dot(theta_i), label=f'Iterasi {i}', alpha=0.5)

# Garis akhir setelah optimasi
plt.plot(X, x.dot(theta_opt), color='black', linewidth=2, label='Final Fit')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Visualisasi Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()

print("Theta hasil akhir:", theta_opt)
```
