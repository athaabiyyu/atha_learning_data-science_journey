##### Exercise 1 : Compute Derivatives of Basic Functions

```python
import sympy as sp

# Define a function
x = sp.symbols('x')

function = x**3 + 2*x**2 + x + 5
# Compute the derivative
derivative = sp.diff(function, x)
print("The derivative of the function is:", derivative)
```

##### Exercise 2 : Compute Gradients

```python
import sympy as sp

x, y = sp.symbols('x y') 
function = x**2 + y**2 + x*y
# Compute the gradient
gradient_x = sp.diff(function, x)
gradient_y = sp.diff(function, y)
print("The gradient of the function is: (", gradient_x, ",", gradient_y, ")")
```

##### Exercise 3 : Implement Gradient Descent for Linear Regression

```python
import numpy as np

def gradient_descent(X, y, theta, lr, iterations):
    m = len(y)
    for _ in range(iterations):
        prediction = np.dot(X, theta)
        error = prediction - y
        gradient = (1/m) * np.dot(X.T, error)
        theta -= lr * gradient
    return theta

X = np.array([[1,1],[1,2],[1,3]])
y = np.array([2,2.5,3.5])
theta = np.array([0.1,0.1])
theta = gradient_descent(X, y, theta, 0.1, 1000)
print(theta)
```
