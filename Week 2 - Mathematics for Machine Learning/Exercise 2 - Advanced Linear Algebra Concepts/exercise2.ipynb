##### Exercise 1 : Create Vectors and Matrices Using Numpy

```python
import numpy as np

A = np.array([[2, 5, 3], [4, 2, 6], [1, 8, 9]])

determinant = np.linalg.det(A)
print("Determinant of the matrix A is:", determinant)

inverse = np.linalg.inv(A)
print("\nInverse of the matrix A is:\n", inverse)
```

##### Exercise 2 : Compute Eigenvalues and Eigenvectors

```python
import numpy as np

A = np.array([[2, 5], [3, 4]])

eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues of the matrix A are:", eigenvalues)
print("\nEigenvectors of the matrix A are:\n", eigenvectors)
```

##### Exercise 3 : Perform Singular Value Decomposition

```python
import numpy as np

A = np.array([[2, 5, -2], [3, 4, 7], [1, 0, 6]])

U, S, VT = np.linalg.svd(A)
print("Left singular vectors (U):\n", U)
print("\nSingular values (S):\n", S)
print("\nRight singular vectors (VT):\n", VT)

# reconstruct
Sigma = np.zeros(A.shape)
np.fill_diagonal(Sigma, S)
reconstructed = U @ Sigma @ VT
print("\nReconstructed matrix A from SVD:\n", reconstructed)
```
