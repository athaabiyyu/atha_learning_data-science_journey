```python
import numpy as np

# buat matrix 3x3
A = np.array([[4, 7, 2],
              [3, 6, 1],
              [2, 5, 3]])

# hitung determinan dan inverse
det_A =  np.linalg.det(A)
print("Determinan A:", det_A)

if det_A != 0:
       inv_A = np.linalg.inv(A)
       print("\nInverse A:\n", inv_A)
else:
       print("\nMatrix A tidak memiliki inverse karena determinannya nol.")
       
# hitung eigenvalues dan eigenvectors
eig_values, eig_vectors = np.linalg.eig(A)
print("\nEigenvalues A:", eig_values)
print("\nEigenvectors A:\n", eig_vectors)

# lakukan pemecahan matrix / SVD
U, S, VT = np.linalg.svd(A)
print("\nU Matrix:\n", U)
print("\nSingular Values:\n", S)
print("\nVT Matrix:\n", VT)

# rekonstruksi matrix dari SVD
Sigma = np.zeros((A.shape[0], A.shape[1]))
np.fill_diagonal(Sigma, S)
A_reconstructed = np.dot(U, np.dot(Sigma, VT))
print("\nReconstructed A from SVD:\n", A_reconstructed)
```
