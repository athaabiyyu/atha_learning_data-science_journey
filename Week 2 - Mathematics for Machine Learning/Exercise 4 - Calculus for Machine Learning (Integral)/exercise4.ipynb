##### Exercise 1 : Calculate Integrals of Simple Functions

```python
import sympy as sp

# Define a function
x = sp.symbols('x')

f = sp.exp(-x)

# compute indefinite integral
indef_integral = sp.integrate(f, x)
print("Indefinite Integral of e^(-x):", indef_integral)

# compute definite integral
def_integral = sp.integrate(f, (x, 0, sp.oo))
print("Definite Integral :", def_integral)
```

##### Exercise 2 : Implement Stochastic Gradient Descent for a Linear Model

```python
import numpy as np

# Data acak (x dan y)
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)

# Tambahkan bias
x_b = np.c_[np.ones((100, 1)), x]

# Fungsi Stochastic Gradient Descent
def stochastic_gradient_descent(x, y, theta, learning_rate, n_epochs):
    m = len(y)
    for epoch in range(n_epochs):
        for i in range(m):
            random_idx = np.random.randint(m)
            xi = x[random_idx:random_idx+1]
            yi = y[random_idx:random_idx+1]
            gradient = 2 * xi.T.dot(xi.dot(theta) - yi)
            theta -= learning_rate * gradient
    return theta

# Inisialisasi parameter
theta_init = np.random.randn(2, 1)
theta_opt = stochastic_gradient_descent(x_b, y, theta_init, learning_rate=0.01, n_epochs=50)

print("Optimized θ (parameters):")
print(theta_opt)
```

##### Exercise 3 : Implement Gradient Descent for Linear Regression

```python
import numpy as np

def gradient_descent(X, y, theta, lr, iterations):
    m = len(y)
    for _ in range(iterations):
        prediction = np.dot(X, theta)
        error = prediction - y
        gradient = (1/m) * np.dot(X.T, error)
        theta -= lr * gradient
    return theta

X = np.array([[1,1],[1,2],[1,3]])
y = np.array([2,2.5,3.5])
theta = np.array([0.1,0.1])
theta = gradient_descent(X, y, theta, 0.1, 1000)
print(theta)
```
