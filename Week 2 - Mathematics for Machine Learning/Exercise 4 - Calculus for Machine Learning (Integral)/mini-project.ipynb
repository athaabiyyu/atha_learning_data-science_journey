```python
import sympy as sp

# definisikan simpol dan fungsi
x = sp.symbols('x')
f = x**2

# hitung integral tak tentu (indefinite)
indefinite_integral = sp.integrate(f, x)
print("Indefinite Integral of x^2:", indefinite_integral)

# hitung integral tentu (definite) dari 0 sampai 2
definite_integral = sp.integrate(f, (x, 0, 2))
print("Definite Integral of x^2 from 0 to 2:", definite_integral)
```

```python
import numpy as np

# Dataset kecil
X = np.array([[1], [2], [3]])
y = np.array([[2], [2.5], [3.5]])

# Tambahkan bias (1 di depan setiap baris)
X_b = np.c_[np.ones((3, 1)), X]

# Inisialisasi parameter
theta = np.random.randn(2, 1)
learning_rate = 0.1
n_epochs = 50

# SGD
for epoch in range(n_epochs):
    for i in range(len(X)):
        random_index = np.random.randint(len(X))
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        theta = theta - learning_rate * gradients

print("Theta hasil SGD:", theta)
```
